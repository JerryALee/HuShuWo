```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def update_points(num):
    point_ani.set_data(x, A[num] * y)
    return point_ani,
 
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

t = np.arange(0, 4*np.pi, 0.1*np.pi)
A = np.sin(t)
 
fig = plt.figure()
point_ani, = plt.plot(x, A[0] * y)
plt.grid(ls="--")
plt.ylim(-1.2, 1.2)
# 开始制作动画
ani = animation.FuncAnimation(fig, update_points, np.arange(0, 40), interval=40, blit=True)

ani.save('sin_test2.gif', writer='imagemagick', fps=10)
# plt.show()
```
